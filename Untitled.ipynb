```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft, fftfreq


#   
file_paths = ['2v.csv', '5v.csv', '8v.csv', '11v.csv', '14v.csv']  #    
colors = ['r', 'g', 'b', 'c', 'm']  #    
labels = ['2V', '5V', '8V', '11V', '14V']


main_title = 'Amp vs Freq'
subtitle = 'gam = 0.1 // 10 kHz vs 9.88 kHz // tau = 0.1'


#    (: 0~5000 Hz)
xlim = (8500, 11500)#     
plt.figure(figsize=(14, 8))


for i, file_path in enumerate(file_paths):
    # CSV   
    data = pd.read_csv(file_path, skiprows=6)
    
    #    
    data.columns = ["Time", "Amplitude"]
    data = data.apply(pd.to_numeric, errors='coerce').dropna()

    #     
    time = data['Time'].values
    amplitude = data['Amplitude'].values

    #   
    N = len(time)

    #  
    T = np.mean(np.diff(time))

    # FFT 
    yf = fft(amplitude)
    xf = fftfreq(N, T)[:N//2]

    #    (    )
    plt.scatter(xf, 2.0/N * np.abs(yf[:N//2]), color=colors[i], s=10)  # s   
    plt.plot(xf, 2.0/N * np.abs(yf[:N//2]), color=colors[i], label=labels[i], linestyle='-', linewidth=1)


# x  
plt.xlim(xlim)


#  
plt.suptitle(main_title, fontsize=16)
plt.title(subtitle, fontsize=12)
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.grid()
plt.legend()
plt.show()
```
