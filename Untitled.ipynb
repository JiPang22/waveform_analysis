```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 데이터 불러오기 및 전처리
data = pd.read_csv('C1Trace00001.csv', names=['Time', 'Voltage'])
data = data[data['Time'] >= 0].reset_index(drop=True)

time_ms = np.linspace(0, 2, 1000)  # 0부터 2ms까지의 가상 시간 데이터
voltage_mV = 100 * np.exp(-0.15 * time_ms) * np.cos(2 * np.pi * 10 * time_ms)  # 감쇠 진동 시뮬레이션 데이터

# 기준 시간 t_0 설정
omega_0 = 10 * 1e3  # 10 kHz = 10,000 rad/s
t_0 = 1 / omega_0   # 기준 시간 (s) -> 0.1 ms

# 무차원화된 시간 계산
dimensionless_time = time_ms / (t_0 * 1000)  # 무차원화된 시간

# 무차원화된 시간 구간 조정 변수
dimensionless_time_limit = 20  # 조정 가능한 무차원화된 시간 구간

# 설정된 무차원화된 시간 구간 이하의 데이터 선택
mask = dimensionless_time <= dimensionless_time_limit
dimensionless_time, voltage_mV = dimensionless_time[mask], voltage_mV[mask]

# 감쇠 계수 및 델타 값 범위 설정 (0.01 단위로 변화)
gam_values = np.arange(0.10, 0.16, 0.01)
delta_values = np.arange(0.10, 0.16, 0.01)

# 서브플롯 생성
fig, axes = plt.subplots(len(gam_values), len(delta_values), figsize=(15, 10))

tlist2 = np.linspace(0, 19, 20)  # 20개의 데이터 포인트

# 반복문을 통한 서브플롯 작성
for i, (gam, delta) in enumerate(np.ndindex(len(gam_values), len(delta_values))):
    gam_val = gam_values[gam]
    delta_val = delta_values[delta]

    # 이론 시뮬레이션 데이터 계산
    ylist = 80 + 63 * np.exp(-gam_val * tlist2 + delta_val)

    # 서브플롯에 원본 데이터 및 시뮬레이션 데이터 플롯
    ax = axes[gam, delta]
    ax.plot(dimensionless_time, voltage_mV, label='Original Data')
    ax.plot(tlist2, ylist, label=f'Simulation (gam={gam_val:.2f}, delta={delta_val:.2f})')
    ax.set_title(f'gam={gam_val:.2f}, delta={delta_val:.2f}')
    ax.grid(True)
    ax.legend()

plt.tight_layout()
plt.show()
```
