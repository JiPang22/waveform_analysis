```python
import numpy as np
import pandas as pd
from PyEMD import EMD
import matplotlib.pyplot as plt
from scipy.signal import hilbert

# 데이터 불러오기 및 전처리
data = pd.read_csv('C1Trace00001.csv', names=['Time', 'Voltage'])
data = data[data['Time'] >= 0].reset_index(drop=True)

time_ms = data['Time'].to_numpy() * 1000  # 밀리초 변환
voltage_mV = data['Voltage'].to_numpy() * 1000  # 밀리볼트 변환

# 기준 시간 t_0 설정
omega_0 = 10 * 1e3  # 10 kHz = 10,000 rad/s
t_0 = 1 / omega_0   # 기준 시간 (s) -> 0.1 ms

# 무차원화된 시간 계산
dimensionless_time = time_ms / (t_0 * 1000)  # 무차원화된 시간

# 무차원화된 시간 구간 조정 변수
dimensionless_time_limit = 100  # 조정 가능한 무차원화된 시간 구간 (예: 40)

# 설정된 무차원화된 시간 구간 이하의 데이터 선택
mask = dimensionless_time <= dimensionless_time_limit
dimensionless_time, voltage_mV = dimensionless_time[mask], voltage_mV[mask]

gam = 0.1   # 감쇠 계수

#imax=linspace(0,len(dimensionless_time))
tmax = np.linspace(0, len(dimensionless_time) - 1, num=len(dimensionless_time))
ylist=[]
for i in range(len(dimensionless_time)):
    y = np.exp(-gam*dimensionless_time[i])+75
    ylist.append(y)


# 원본 데이터 플롯
plt.plot(dimensionless_time, voltage_mV, label=r'Original Data($\omega_0$ = 10 kHz)')
# 이론 시뮬레이션 플롯
plt.plot(dimensionless_time, ylist, '--', label='Envelope', color='red')
plt.grid(True)
plt.show()
```
