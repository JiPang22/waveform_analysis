```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일 경로 설정
csv_file_path = 'C1Trace00007.csv'

# CSV 파일 읽기 (필요한 데이터가 6번째 줄부터 시작), 컬럼 이름 지정
data = pd.read_csv(csv_file_path, skiprows=5, header=None, names=['Time', 'Ampl'])

# 데이터 확인 (첫 몇 줄)
data.head()

# 'Time'과 'Ampl' 컬럼 추출
time_data = data['Time']
amplitude_data = data['Ampl']

# 플롯 생성
plt.figure(figsize=(10, 6))
plt.plot(time_data, amplitude_data, label='Waveform')

# 플롯 설정
plt.title('Waveform Data')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)

# 플롯 표시
plt.show()
```

```python
import numpy as np

# 푸리에 변환 수행
amplitude_fft = np.fft.fft(amplitude_data)
frequency = np.fft.fftfreq(len(amplitude_data), (time_data[1] - time_data[0]))

# 푸리에 변환 결과 시각화
plt.figure(figsize=(10, 6))
plt.plot(frequency, np.abs(amplitude_fft), label='FFT of Waveform')
plt.title('Fourier Transform of Waveform Data')
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)

# 플롯 표시
plt.show()
```
