```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob
import os

#    (: 0 ~ 500 Hz)
min_freq = 1600 - 700  #  
max_freq = 1600 + 1000  #  

#    .txt  
file_paths = glob.glob('*.txt')

#  FFT   
frequency_dict = {}

for file_path in file_paths:
    #      
    data = pd.read_csv(file_path, skiprows=8, header=None)
    
    #    ,     
    time = data.iloc[:, 0]
    amplitude = data.iloc[:, 1]

    # FFT 
    n = len(amplitude)  #   
    dt = time[1] - time[0]  #   (  )
    frequencies = np.fft.fftfreq(n, d=dt)
    fft_values = np.fft.fft(amplitude)

    #    FFT  
    positive_frequencies = frequencies[:n // 2]
    positive_fft_values = np.abs(fft_values[:n // 2])

    #   
    mask = (positive_frequencies >= min_freq) & (positive_frequencies <= max_freq)
    filtered_frequencies = positive_frequencies[mask]
    filtered_fft_values = positive_fft_values[mask]

    #     
    for freq, amp in zip(filtered_frequencies, filtered_fft_values):
        if freq not in frequency_dict:
            frequency_dict[freq] = []
        frequency_dict[freq].append(amp)

#        
plt.figure(figsize=(12, 8))
plt.boxplot([frequency_dict[freq] for freq in sorted(frequency_dict.keys())], 
            patch_artist=True, showmeans=True)

# x   
plt.xticks(ticks=range(1, len(frequency_dict) + 1), labels=[f'{freq:.2f}' for freq in sorted(frequency_dict.keys())], rotation=45)

#  
plt.title("Box Plot of Amplitude Distribution by Frequency")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob
import os

#   
min_freq = 1600 - 700  #  
max_freq = 1600 + 1000  #  

#    .txt  
file_paths = glob.glob('*.txt')

#    
frequency_means = {}

for file_path in file_paths:
    #      
    data = pd.read_csv(file_path, skiprows=8, header=None)
    
    #    ,     
    time = data.iloc[:, 0]
    amplitude = data.iloc[:, 1]

    # FFT 
    n = len(amplitude)  #   
    dt = time[1] - time[0]  #   (  )
    frequencies = np.fft.fftfreq(n, d=dt)
    fft_values = np.fft.fft(amplitude)

    #    FFT  
    positive_frequencies = frequencies[:n // 2]
    positive_fft_values = np.abs(fft_values[:n // 2])

    #   
    mask = (positive_frequencies >= min_freq) & (positive_frequencies <= max_freq)
    filtered_frequencies = positive_frequencies[mask]
    filtered_fft_values = positive_fft_values[mask]

    #     
    for freq, amp in zip(filtered_frequencies, filtered_fft_values):
        if freq not in frequency_means:
            frequency_means[freq] = []
        frequency_means[freq].append(amp)

#     
mean_values = {freq: np.mean(amps) for freq, amps in frequency_means.items()}

#   
sorted_frequencies = sorted(mean_values.keys())
sorted_mean_values = [mean_values[freq] for freq in sorted_frequencies]

#    
plt.figure(figsize=(12, 8))
plt.plot(sorted_frequencies, sorted_mean_values, 'go-', label='Mean Amplitude')

#  
plt.title("Mean Amplitude by Frequency (Sorted)")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Mean Amplitude")
plt.grid(True)
plt.legend()
plt.show()
```
