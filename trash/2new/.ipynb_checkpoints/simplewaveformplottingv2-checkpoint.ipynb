```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the file
file_path = 'C1Trace00000.txt'

# Load the relevant data by skipping the header rows
data = pd.read_csv(file_path, skiprows=8)

# Extract the time and amplitude values
time = data["Time"]
amplitude = data["Ampl"]

# Perform FFT
n = len(amplitude)  # Number of data points
dt = time[1] - time[0]  # Time interval (assuming uniform spacing)
frequencies = np.fft.fftfreq(n, d=dt)
fft_values = np.fft.fft(amplitude)

# Only keep the positive frequencies and their corresponding FFT values
positive_frequencies = frequencies[:n // 2]
positive_fft_values = np.abs(fft_values[:n // 2])

# Plot the FFT result
plt.figure(figsize=(10, 6))
plt.plot(positive_frequencies, positive_fft_values, color='red')
plt.title("Frequency Spectrum")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob
import os

#    .txt  
file_paths = glob.glob('*.txt')

#    
for file_path in file_paths:
    #      
    data = pd.read_csv(file_path, skiprows=8)

    #     
    time = data["Time"]
    amplitude = data["Ampl"]

    # FFT 
    n = len(amplitude)  #   
    dt = time[1] - time[0]  #   (  )
    frequencies = np.fft.fftfreq(n, d=dt)
    fft_values = np.fft.fft(amplitude)

    #    FFT  
    positive_frequencies = frequencies[:n // 2]
    positive_fft_values = np.abs(fft_values[:n // 2])

    #    
    file_name = os.path.splitext(os.path.basename(file_path))[0]

    #     ( )
    plt.figure(figsize=(10, 6))
    plt.plot(time, amplitude, 'o', color='blue', markersize=2)
    plt.title(f"Time Series Data ({file_name})")
    plt.xlabel("Time (s)")
    plt.ylabel("Amplitude")
    plt.grid(True)
    plt.show()

    # FFT   ( )
    plt.figure(figsize=(10, 6))
    plt.plot(positive_frequencies, positive_fft_values, 'o', color='red', markersize=2)
    plt.title(f"Frequency Spectrum ({file_name})")
    plt.xlabel("Frequency (Hz)")
    plt.ylabel("Amplitude")
    plt.grid(True)
    plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob
import os

#    .txt  
file_paths = glob.glob('*.txt')

# FFT   
plt.figure(figsize=(10, 6))

for file_path in file_paths:
    #      
    data = pd.read_csv(file_path, skiprows=8, header=None)
    
    #    ,     
    time = data.iloc[:, 0]
    amplitude = data.iloc[:, 1]

    # FFT 
    n = len(amplitude)  #   
    dt = time[1] - time[0]  #   (  )
    frequencies = np.fft.fftfreq(n, d=dt)
    fft_values = np.fft.fft(amplitude)

    #    FFT  
    positive_frequencies = frequencies[:n // 2]
    positive_fft_values = np.abs(fft_values[:n // 2])

    #      
    file_name = os.path.splitext(os.path.basename(file_path))[0]

    # FFT    ( )
    plt.plot(positive_frequencies, positive_fft_values, 'o-', markersize=2, label=file_name)

#  
plt.title("Overlapped Frequency Spectrum")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob
import os

#    (: 0 ~ 500 Hz)
min_freq = 1600-700  #  
max_freq = 1600+1000  #  

#    .txt  
file_paths = glob.glob('*.txt')

# FFT   
plt.figure(figsize=(10, 6))

for file_path in file_paths:
    #      
    data = pd.read_csv(file_path, skiprows=8, header=None)
    
    #    ,     
    time = data.iloc[:, 0]
    amplitude = data.iloc[:, 1]

    # FFT 
    n = len(amplitude)  #   
    dt = time[1] - time[0]  #   (  )
    frequencies = np.fft.fftfreq(n, d=dt)
    fft_values = np.fft.fft(amplitude)

    #    FFT  
    positive_frequencies = frequencies[:n // 2]
    positive_fft_values = np.abs(fft_values[:n // 2])

    #   
    mask = (positive_frequencies >= min_freq) & (positive_frequencies <= max_freq)
    filtered_frequencies = positive_frequencies[mask]
    filtered_fft_values = positive_fft_values[mask]

    #      
    file_name = os.path.splitext(os.path.basename(file_path))[0]

    # FFT    ( )
    plt.plot(filtered_frequencies, filtered_fft_values, 'o-', markersize=2, label=file_name)

#  
plt.title("Overlapped Frequency Spectrum")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob
import os

#    (: 0 ~ 500 Hz)
min_freq = 1600 - 700  #  
max_freq = 1600 + 1000  #  

#    .txt  
file_paths = glob.glob('*.txt')

# FFT   
plt.figure(figsize=(10, 6))

for file_path in file_paths:
    #      
    data = pd.read_csv(file_path, skiprows=8, header=None)
    
    #    ,     
    time = data.iloc[:, 0]
    amplitude = data.iloc[:, 1]

    # FFT 
    n = len(amplitude)  #   
    dt = time[1] - time[0]  #   (  )
    frequencies = np.fft.fftfreq(n, d=dt)
    fft_values = np.fft.fft(amplitude)

    #    FFT  
    positive_frequencies = frequencies[:n // 2]
    positive_fft_values = np.abs(fft_values[:n // 2])

    #   
    mask = (positive_frequencies >= min_freq) & (positive_frequencies <= max_freq)
    filtered_frequencies = positive_frequencies[mask]
    filtered_fft_values = positive_fft_values[mask]

    #       
    dominant_index = np.argmax(filtered_fft_values)
    dominant_frequency = filtered_frequencies[dominant_index]
    dominant_value = filtered_fft_values[dominant_index]

    #      
    file_name = os.path.splitext(os.path.basename(file_path))[0]

    #    ( )
    plt.plot(dominant_frequency, dominant_value, 'o', markersize=8, label=file_name)

#  
plt.title("Dominant Frequency Points")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.legend()
plt.show()
```
