```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob
import os

#    (: 0 ~ 500 Hz)
min_freq = 1600-700  #  
max_freq = 1600+1000  #  

#    .txt  
file_paths = glob.glob('*.txt')

# FFT   
plt.figure(figsize=(10, 6))

for file_path in file_paths:
    #      
    data = pd.read_csv(file_path, skiprows=8, header=None)
    
    #    ,     
    time = data.iloc[:, 0]
    amplitude = data.iloc[:, 1]

    # FFT 
    n = len(amplitude)  #   
    dt = time[1] - time[0]  #   (  )
    frequencies = np.fft.fftfreq(n, d=dt)
    fft_values = np.fft.fft(amplitude)

    #    FFT  
    positive_frequencies = frequencies[:n // 2]
    positive_fft_values = np.abs(fft_values[:n // 2])

    #   
    mask = (positive_frequencies >= min_freq) & (positive_frequencies <= max_freq)
    filtered_frequencies = positive_frequencies[mask]
    filtered_fft_values = positive_fft_values[mask]

    #      
    file_name = os.path.splitext(os.path.basename(file_path))[0]

    # FFT    ( )
    plt.plot(filtered_frequencies, filtered_fft_values, 'o-', markersize=2, label=file_name)

#  
plt.title("Overlapped Frequency Spectrum")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob
import os

#    (: 0 ~ 500 Hz)
min_freq = 1600 - 700  #  
max_freq = 1600 + 1000  #  

#    .txt  
file_paths = glob.glob('*.txt')

# FFT   
all_fft_values = []

for file_path in file_paths:
    #      
    data = pd.read_csv(file_path, skiprows=8, header=None)
    
    #    ,     
    time = data.iloc[:, 0]
    amplitude = data.iloc[:, 1]

    # FFT 
    n = len(amplitude)  #   
    dt = time[1] - time[0]  #   (  )
    frequencies = np.fft.fftfreq(n, d=dt)
    fft_values = np.fft.fft(amplitude)

    #    FFT  
    positive_frequencies = frequencies[:n // 2]
    positive_fft_values = np.abs(fft_values[:n // 2])

    #   
    mask = (positive_frequencies >= min_freq) & (positive_frequencies <= max_freq)
    filtered_fft_values = positive_fft_values[mask]

    # FFT   
    all_fft_values.append(filtered_fft_values)

#   
plt.figure(figsize=(10, 6))
plt.boxplot(all_fft_values, vert=False, patch_artist=True, showmeans=True)

#  
plt.title("Box Plot of Frequency Spectrum")
plt.xlabel("Amplitude")
plt.ylabel("Frequency Range Segments")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob
import os

#    (: 0 ~ 500 Hz)
min_freq = 1600 - 700  #  
max_freq = 1600 + 1000  #  

#    .txt  
file_paths = glob.glob('*.txt')

#  FFT   
frequency_dict = {}

for file_path in file_paths:
    #      
    data = pd.read_csv(file_path, skiprows=8, header=None)
    
    #    ,     
    time = data.iloc[:, 0]
    amplitude = data.iloc[:, 1]

    # FFT 
    n = len(amplitude)  #   
    dt = time[1] - time[0]  #   (  )
    frequencies = np.fft.fftfreq(n, d=dt)
    fft_values = np.fft.fft(amplitude)

    #    FFT  
    positive_frequencies = frequencies[:n // 2]
    positive_fft_values = np.abs(fft_values[:n // 2])

    #   
    mask = (positive_frequencies >= min_freq) & (positive_frequencies <= max_freq)
    filtered_frequencies = positive_frequencies[mask]
    filtered_fft_values = positive_fft_values[mask]

    #     
    for freq, amp in zip(filtered_frequencies, filtered_fft_values):
        if freq not in frequency_dict:
            frequency_dict[freq] = []
        frequency_dict[freq].append(amp)

#        
plt.figure(figsize=(12, 8))
plt.boxplot([frequency_dict[freq] for freq in sorted(frequency_dict.keys())], 
            patch_artist=True, showmeans=True)

# x   
plt.xticks(ticks=range(1, len(frequency_dict) + 1), labels=[f'{freq:.2f}' for freq in sorted(frequency_dict.keys())], rotation=45)

#  
plt.title("Box Plot of Amplitude Distribution by Frequency")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.tight_layout()
plt.show()
```
