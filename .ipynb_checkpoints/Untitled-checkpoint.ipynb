```python
import numpy as np
import pandas as pd
from PyEMD import EMD
import matplotlib.pyplot as plt
from scipy.signal import hilbert

# 데이터 불러오기 및 전처리
data = pd.read_csv('C1Trace00001.csv', names=['Time', 'Voltage'])
data = data[data['Time'] >= 0].reset_index(drop=True)

time_ms = data['Time'].to_numpy() * 1000  # 밀리초 변환
voltage_mV = data['Voltage'].to_numpy() * 1000  # 밀리볼트 변환

# 기준 시간 t_0 설정
omega_0 = 10 * 1e3  # 10 kHz = 10,000 rad/s
t_0 = 1 / omega_0   # 기준 시간 (s) -> 0.1 ms

# 무차원화된 시간 계산
dimensionless_time = time_ms / (t_0 * 1000)  # 무차원화된 시간

# 무차원화된 시간 구간 조정 변수
dimensionless_time_limit = 100  # 조정 가능한 무차원화된 시간 구간 (예: 40)

# 설정된 무차원화된 시간 구간 이하의 데이터 선택
mask = dimensionless_time <= dimensionless_time_limit
dimensionless_time, voltage_mV = dimensionless_time[mask], voltage_mV[mask]

# EMD 분석
emd = EMD()
imfs = emd.emd(voltage_mV)

# IMF 11의 데이터 추출
imf_11 = imfs[10]  # IMF 11 (0-based index에서 10은 11번째 IMF)

# IMF 11의 엔벨로프 계산
analytic_signal = hilbert(imf_11)
amplitude_envelope = np.abs(analytic_signal)

# 시스템 조건 설정 (이론 시뮬레이션)
dt = 1.e-2  # 시간 스텝
gam = 0.1   # 감쇠 계수
imax = 10000  # 시뮬레이션의 최대 반복 횟수 (100까지 설정)

# 시뮬레이션 시간 설정
t_sim = np.linspace(0, 100, imax)  # 시간 범위를 100으로 설정
x_sim = np.zeros(imax)
y_sim = np.ones(imax)

# 시스템 통합 (시뮬레이션 진행)
for i in range(1, imax):
    dx = y_sim[i - 1]
    dy = -gam * y_sim[i - 1] - x_sim[i - 1]
    x_sim[i] = x_sim[i - 1] + dx * dt
    y_sim[i] = y_sim[i - 1] + dy * dt

# 이론 시뮬레이션 데이터의 힐버트 변환을 사용한 엔벨로프 계산
analytic_signal_sim = hilbert(x_sim)
envelope_sim = np.abs(analytic_signal_sim)

# 플롯 생성
fig, axes = plt.subplots(4, 1, figsize=(12, 16), dpi=300)

# 원본 데이터 플롯
plt.plot(dimensionless_time, voltage_mV, label=r'Original Data($\omega_0$ = 10 kHz)')

# 이론 시뮬레이션 플롯
plt.plot(t_sim, envelope_sim, '--', label='Envelope', color='red')


plt.set(xlabel=r"Time ($\omega_0$t)", ylabel='Amplitude (mV)')
plt.legend(loc='upper right')


plt2].set(title='Theoretical Damped Oscillator', xlabel=r"Time ($\omega_0$t)", ylabel='Amplitude')
plt2].legend(loc='upper right')
plt2].grid(True)

```
