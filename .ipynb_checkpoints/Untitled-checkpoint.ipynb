```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일 경로 설정
csv_file_path = '2v.csv'

# CSV 파일 읽기 (필요한 데이터가 6번째 줄부터 시작), 컬럼 이름 지정
data = pd.read_csv(csv_file_path, skiprows=5, header=None, names=['Time', 'Ampl'])

# 데이터 확인 (첫 몇 줄)
data.head()

# 'Time'과 'Ampl' 컬럼 추출
time_data = data['Time']
amplitude_data = data['Ampl']

# 플롯 생성
plt.figure(figsize=(10, 6))
plt.plot(time_data, amplitude_data, label='Waveform')

# 플롯 설정
plt.title('Waveform Data')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)

# 플롯 표시
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft, fftfreq


# 파일 경로 리스트
file_paths = ['2v.csv', '5v.csv', '8v.csv', '11v.csv', '14v.csv']  # 파일 경로를 여기에 입력하세요
colors = ['r', 'g', 'b', 'c', 'm']  # 각 데이터의 플롯 색상
labels = ['2V', '5V', '8V', '11V', '14V']


main_title = 'Amp vs Freq'
subtitle = 'gam = 0.1 // 10 kHz vs 9.88 kHz // tau = 0.1'


# 가로축 범위 설정 (예: 0~5000 Hz)
xlim = (8500, 11500)# 필요에 따라 원하는 범위로 설정
plt.figure(figsize=(14, 8))


for i, file_path in enumerate(file_paths):
    # CSV 파일에서 데이터 로드
    data = pd.read_csv(file_path, skiprows=6)
    
    # 숫자형 데이터만 남기기 
    data.columns = ["Time", "Amplitude"]
    data = data.apply(pd.to_numeric, errors='coerce').dropna()

    # 시간 및 진폭 데이터 추출
    time = data['Time'].values
    amplitude = data['Amplitude'].values

    # 데이터 포인트 개수
    N = len(time)

    # 샘플 간격
    T = np.mean(np.diff(time))

    # FFT 수행
    yf = fft(amplitude)
    xf = fftfreq(N, T)[:N//2]

    # 결과 점 플롯 (각 데이터 다른 색상과 라벨)
    plt.scatter(xf, 2.0/N * np.abs(yf[:N//2]), color=colors[i], s=10)  # s는 점의 크기 조절
    plt.plot(xf, 2.0/N * np.abs(yf[:N//2]), color=colors[i], label=labels[i], linestyle='-', linewidth=1)


# x축 범위 적용
plt.xlim(xlim)


# 플롯 설정
plt.suptitle(main_title, fontsize=16)
plt.title(subtitle, fontsize=12)
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.grid()
plt.legend()
plt.show()
```
